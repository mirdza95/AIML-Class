```python
iris_without_species = df.drop('species', axis= 1)
iris_without_species.head()
```

```python
# instead

species2 = df.drop(['sepal_length','sepal_width','petal_length','petal_width'], axis=1)

species2.head()

type(species2)
```

```python
iris_without_species.head()
```

```python
# calculated column

sepalx = iris_without_species['sepal_length'] * iris_without_species['sepal_width']

sepalx.head()
```

```python
iris_without_species.head()
```

```python
iris_without_species['sepal_area'] = sepalx

iris_without_species.head()
```

```python
# petal area!!!

idreesx = iris_without_species['petal_length'] * iris_without_species['petal_width']

idreesx.head()
```

```python
iris_without_species['petal_area'] = idreesx

iris_without_species.head()

# only part of iris_without_species value of sepal_area is > 14
```

```python
# filter out parts of dataframes using pandas

largeFlowers = iris_without_species[iris_without_species['sepal_area'] > 18 ]

largeFlowers.head(10)

# how many large flowers we have in iris dataset\
# how many flowers have a sepal area of exactly 16

largeFlowers.info()
```

```python
largeFlowers16 = iris_without_species[iris_without_species['sepal_area'] == 16 ]

largeFlowers16.info()
```

```python
largeFlowers
```

```python
largeFlowers.to_csv('largeflowers.csv')
```

```python
# sorting values by something

sortedLargeFlowers = largeFlowers.sort_values(by='sepal_area', ascending=False)

sortedLargeFlowers
```
